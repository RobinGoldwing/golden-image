```python
# Cargamos librerías
import numpy as np
import cv2
import skimage as ski;
from skimage import transform;
from ipywidgets import interact
```

### Cargamos imagen

```python
# Cargamos imagen

# Desde Un archivo
original = cv2.imread("images/coche.jpg")
original = cv2.cvtColor(original,cv2.COLOR_BGR2RGB)
# Mostramos imágen
ski.io.imshow(original)
```

### Convertimos a escala de grises

```python
# Convertimos a escala de grises
gris = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
# Mostramos imagen
ski.io.imshow(gris)
```

### Ajusta el filtro Gaussiano

```python
# Creamos slider interactivo
@interact(x=(1, 99, 2))
# Función que ajusta el filtro Gaussiano
def ajusta_filtro(x):
    global gauss
    
    gauss = cv2.GaussianBlur(gris, (x,x), 0)
    ski.io.imshow(gauss)
```
