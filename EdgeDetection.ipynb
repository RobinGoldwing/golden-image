## Creación de máscara de contornos en fondo blanco
---


#### Declaración de librerias

```python
%matplotlib inline

import matplotlib as mplt;

from matplotlib import pyplot as plt;

import skimage as ski;

from ipywidgets import interact


```

#### Carga de imagen, y muestra del ejemplo

```python
# GLOBO
# img = ski.io.imread("images/globo.jpg", 0);

# COCHE
# img = ski.io.imread("images/coche.jpg");

# ZAPATOS
img = ski.io.imread("images/zapatos.jpg");


ski.io.imshow(img)
```

#### Creación de máscara por tonos blancos

```python
img_ = img.copy()

masked = img

@interact(x=(0, 255))

def double_number(x):
    global masked
    w_mask = img_ > x
    masked = img * w_mask
    ski.io.imshow(masked)
```

#### Resta del fondo blanco a la imagen origina, y presentación de resultados

```python

diff = (img_ - masked)

ski.io.imshow(diff)

#img_[w_mask]=[255,255,0]
#ski.io.imshow(img_)
```

## Detección de contornos de objetos
---

### Importamos librerías

```python
import cv2
from cv2 import Canny
```

### Cargamos imagen 

```python
# Globo
# imagen = cv2.imread("images/coche.jpg")
# Zapatos
# imagen = cv2.imread("images/coche.jpg")
# COCHE
imagen = cv2.imread("images/coche.jpg")
imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)


plt.figure(figsize=(10,10))
plt.imshow(imagen)
plt.axis("off")
plt.show()
```

### Convertimos la imagen en escala de grises

```python
imagen_gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
```

```python

@interact(x=(0, 500),y=(0, 500))

def ajusta_contorno(x, y):
    global contorno
    contorno = cv2.Canny(imagen_gris, threshold1=x, threshold2=y)
    plt.figure(figsize=(10,10))
    plt.imshow(contorno,cmap="gray")
    plt.axis("off")
    plt.show()
```
